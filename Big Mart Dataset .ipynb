## Data Processing

```python
print('Skewness: %f' % dataset['Item_Outlet_Sales'].skew())
print('Kurtsis: %f' %dataset['Item_Outlet_Sales'].kurt())
```

```python
def remove_outlier_boxplot(dataset,col):
    q3 = np.quantile(dataset[col], .75)
    q1 = np.quantile(dataset[col], .25)
    iqr = q3-q1
    print("iqr is ..", iqr)
    global outlier_higher_side
    global outlier_lower_side
    outlier_higher_side = [x for x in dataset[col] if (x >  (q3 + 1.5*iqr))]
    outlier_lower_side = [x for x in dataset[col] if (x <  (q1 - 1.5*iqr))]
    
```

```python
remove_outlier_boxplot(dataset,'Item_Outlet_Sales')
```

```python
outlier_lower_side
```

```python
len(outlier_higher_side)
```

```python
def remove_outlier_boxplot_reverse(dataset,col):
    q3 = np.quantile(dataset[col], .75)
    q1 = np.quantile(dataset[col], .25)
    iqr = q3-q1
    global outlier_free_list
    global outlier_free_dataset
    print("iqr is ..", iqr)
    #global outlier_free_df
    outlier_free_list = [ x for x in dataset[col] if ((x <  (q3 + 1.5*iqr)) & (x >  (q1 - 1.5*iqr)))]
    outlier_free_dataset = dataset.loc[dataset[col].isin(outlier_free_list)] ; print(outlier_free_dataset.shape);return outlier_free_dataset
```

```python
dataset = remove_outlier_boxplot_reverse(dataset,'Item_Outlet_Sales')
```

```python
def remove_outliers (dataset,col,k=2):
    mean=dataset[col].mean()
    global df1
    sd = dataset[col].std()
    final_list = [x for x in dataset[col] if (x > mean -k *sd)]
    final_list = [x for x in dataset[col] if (x > mean + k *sd)]
    df1 = dataset.loc[dataset[col].isin(final_list)] ; print(dataset.shape); return df1
```

```python
remove_outliers(dataset,'Item_Outlet_Sales' ,k=2)
```
